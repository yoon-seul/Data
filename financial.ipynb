```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
nd1 = pd.read_excel("4-1-1. 국가채무(D1).xlsx")
nd2 = pd.read_excel("4-1-2. 월별 국가채무(중앙정부).xlsx")
```

```python
nd1.isna().sum()
```

```python
nd1.fillna(nd1.iloc[:,-1].mean(),inplace=True)
```

```python
plt.rcParams['font.family']='NanumGothic'
```

```python
plt.plot(nd1.groupby('기준년도').mean()['국가채무 GDP대비(%)'])
plt.title('기준년도 - 국가채무 GDP대비(%)',fontsize=15)
plt.xlabel('국가채무 GDP대비(%)',fontsize=15)
plt.ylabel('기준년도',fontsize=15)
```

```python
nd2.isna().sum()
```

```python
plt.plot(nd2.groupby('기준월').mean()['국가채무총계(조원)'])
plt.title('기준월-국가채무총계(조원)',fontsize=15)
plt.xlabel('기준월', fontsize=15)
plt.ylabel('국가채무총계(조원)',fontsize=15)
```
